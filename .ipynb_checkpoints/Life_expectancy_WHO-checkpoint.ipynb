```python
from sklearn.preprocessing import LabelEncoder
cate_col=["Country","Status"]
encode=dict()
for i in cate_col:
    label=LabelEncoder()
    df[i]=label.fit_transform(df[i])
    encode[i]=label
```

```python
feature_columns = list(df.columns)
label_column = "Life expectancy "
df = df[feature_columns].dropna(subset=[label_column])
```

```python
df.isnull().sum()
```

```python
def train(df,model_callback):
    from sklearn.metrics import mean_absolute_error
    from sklearn.model_selection import GroupKFold
    
    kfold = GroupKFold(5)
    models, maes = [], []
    
    for i,(train_ind,valid_ind) in enumerate(kfold.split(df,groups=df["Country"])):
        X_train,y_train=df.iloc[train_ind][feature_columns],df.iloc[train_ind][label_column]
        X_test,y_test=df.iloc[valid_ind][feature_columns],df.iloc[valid_ind][label_column]
        
        model=model_callback(X_train,y_train,X_test,y_test)
        models.append(model)
        
        #maes=np.append(mean_absolute_error(y_test,model.predict(X_test)))
        maes = np.append(maes, [mean_absolute_error(y_test, model.predict(X_test))])

        print(len(train_ind),len(valid_ind))
    
    avg_mae = np.mean(maes)
    print(f"Average MAE: {avg_mae:.2f}")
    return models, {"mae": avg_mae}
```

```python
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from IPython.display import Markdown
```

```python
def catboost_callback(X_train, y_train, X_val, y_val):
    model = CatBoostRegressor(
        iterations=1000,
        depth=8,
        task_type='CPU',
        use_best_model=True,
        eval_metric='MAE',
        early_stopping_rounds=100,
        learning_rate=0.03,
        border_count=32,
        l2_leaf_reg=3,
        verbose=1000
    ).fit(X_train, y_train, eval_set=(X_val, y_val))

    feature_importance_df = pd.DataFrame({
        "Feature": feature_columns,
        "Feature Importance": model.feature_importances_
    }).sort_values(by="Feature Importance", ascending=False)

    plt.figure(figsize=(12, 8))
    sns.barplot(x=feature_importance_df["Feature Importance"], y=feature_importance_df["Feature"])
    plt.title("CatBoost Feature Importance")
    plt.show()

    return model

cat_models, cat_metrics = train(df, catboost_callback)
```
